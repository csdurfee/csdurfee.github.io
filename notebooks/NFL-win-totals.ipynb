calculating all pairwise correlations (there is probably a better way of doing this)

```python
pairwise_corrs = {}

min_correl = 1
min_correl_name = None

for source in sources:
    for source2 in sources:
        if source != source2:
            this_rank = power_rankings[['Rank', source]].set_index(source)["Rank"]
            other_rank = power_rankings[['Rank', source2]].set_index(source2)["Rank"]
            this_corr = this_rank.corr(other_rank)
            pairwise_corrs[(source, source2)] = this_corr
            if this_corr < min_correl:
                min_correl = this_corr
                min_correl_name = (source, source2)
pairwise_corrs
```
